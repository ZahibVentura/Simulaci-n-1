```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def fun(n=50):
  Lista = []
  for i in range(n):
    U = rd.uniform(0,1)
    x = (U**(0.135)-(1-U)**(0.135))/(0.1975)
    Lista.append(x)
  return Lista

Iteraciones=int(input("Digite las iteraciones: "))

G = fun(Iteraciones)
plt.hist(G,bins=1000)
plt.show()
#plt.plot(G)
#plt.show()
```

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def Weibull(alphita=1,n=50):
  Lista = []
  for i in range(n):
    U = rd.uniform(0,1)
    x = (-np.log(U))**(1/alphita)
    Lista.append(x)
  return Lista


itera = int(input("Ingerse el numero de iteraciones: "))
alp = float(input("Ingrese alpha: "))
H =Weibull(alp,itera)

plt.hist(H,bins=1000)
plt.show()
```

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def Gama(n=50):
  Lista = []
  for i in range(n):
    U = rd.uniform(0,1)
    x = -np.log((-np.log(U)))
    Lista.append(x)
  return Lista


itera = int(input("Ingerse el numero de iteraciones: "))
#alp = float(input("Ingrese alpha: "))
Z = Gama(itera)

plt.hist(Z,bins=1000)
plt.show()
```

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import math

def Cuchy(n=50):
  Lista = []
  for i in range(n):
    U = rd.uniform(0,1)
    x = math.tan(math.pi*U)
    Lista.append(x)
  return Lista

#r=(range(-10,10,1))
itera = int(input("Ingerse el numero de iteraciones: "))
#alp = float(input("Ingrese alpha: "))
C =Cuchy(itera)

plt.hist(C,bins=1000)
plt.show()
```

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def Laplace(n=50):
  Lista = []
  for i in range(n):
    U = rd.uniform(0,1)
    x = -np.log((1-U)/U)
    Lista.append(x)
  return Lista


itera = int(input("Ingerse el numero de iteraciones: "))
#alp = float(input("Ingrese alpha: "))
L =Laplace(itera)

plt.hist(L,bins=1000)
plt.show()
```
