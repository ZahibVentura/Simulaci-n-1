```python
import random as rnd
import matplotlib.pyplot as plt
No_veces = int(input('ingrese cuántos lanzamientos: '))
a=0
b=0
c=0
d=0
e=0
f=0
resultado = []
probabilidad = []
Lista = []
for i in range(No_veces):
  Lista.append(i)
  x=rnd.random()
  if 0<=x<0.1666:
    a+=1
    proba_1 = a/(i+1)
    probabilidad.append(proba_1)
    resultado.append(1)
  elif 0.1666<=x<0.3333:
    b+=1
    proba_2 = b/(i+1)
    resultado.append(2)
  elif 0.3333<=x<0.5000:
    c+=1
    proba_3 = c/(i+1)
    resultado.append(3)
  elif 0.5000<=x<0.6666:
    e+=1
    proba_4 = d/(i+1)
    resultado.append(4)
  elif 0.6666<=x<0.8333:
    e+=1
    proba_5 = e/(i+1)
    resultado.append(5)
  elif 0.8333<=x<1.0000:
    f+=1
    proba_6 = f/(i+1)
    resultado.append(6)
print(probabilidad)
plt.plot(probabilidad)
```
