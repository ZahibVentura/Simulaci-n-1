```python
import random as rd
import matplotlib.pyplot as plt   #importamos todas las librerias a utilizar
import numpy as np
```

```python
#La siguiente parte del codigo grafica la función la cual se intenta simular

#Puesto que es una funcion a trozos se define el intervalo 0<=X<1 de para la primera parte de la función
a = 0
b = 1
X = np.linspace(a, b, 50)
Y = X/3   #primer trozo de la función

c = 1   #intervalo 1<=X<3
d = 3
V = np.linspace(c,d, 50)
W = 1/3 + 0*V  #segundo trozo de la función

e = 3   #intervalo 3<=X<=4
f = 4
P = np.linspace(e,f, 50)
O = (4-P)/3  #tercer trozo de la función

x = [0,1,3,4]
y = [0.0,0.333,0.333,0]

plt.figure()

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})') #se muestran los puntos donde hay la intersección de las funciones y extremos

plt.plot(X, Y, 'b')   #se grafican ambas funciones por partes (son tres partes)
plt.plot(V, W, 'b')
plt.plot(P, O, 'b')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')  #titulos de ejes
plt.grid()
plt.show()
```

```python
#La siguiente parte del codigo grafica la función acumulativa

X = np.linspace(a, b, 50)
Y = (X**2)/6     #esta es la función acumulativa del primer trozo de la función

V = np.linspace(c,d, 50)
W = (2*V -1)/6   #esta es la función acumulativa del segundo trozo de la función

P = np.linspace(e,f, 50)
O = (-P**2 +8*P -10)/6   #esta es la función acumulativa del tercer trozo de la función

x = [0,1,3,4]
y = [0.0,0.166,0.833,1]

plt.figure()

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})')  #se muestran los puntos donde hay la intersección de las funciones acumulativas y extremos

plt.plot(X, Y, 'k')
plt.plot(V, W, 'k')  #se grafican ambas funciones acumulativas por partes (son tres partes)
plt.plot(P, O, 'k')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')  #titulos a los ejes
plt.grid()
plt.show()
```

```python
#La función de la grafica es una f.d.p
#Se comprobó que la integral total da como resultado 1
#además la función siempre es positiva para todo X en el intervalo [0,4]
#se aplica el método de la transformada inversa para su simulación
#tomando a:
#            |  (6*u)^0.5    si   u<= 1/6
#            |
#  F-1 (u) = |  (6*u+1)/2     si   1/6 < u <= 5/6
#            |
#            |  4-(6-6*u)^0.5  si  u > 5/6

#El siguiente código muestra la simulación de dicha función

def funcion(N=50): #Definimos una función que regresa una lista de valores x
  Lista = []
  for i in range(N):
    u = rd.uniform(0,1) #variable uniforme de 0 a 1

    if u <= (1/6):   #primera condición
      f = (6*u)**0.5
    elif (1/6) < u <= (5/6):   #segunda condición
      f = (6*u+1)/2
    else:                #ultima condición
      f = 4-(6-6*u)**0.5
    Lista.append(f)            #guardamos los valores en una lista
  return Lista # se regresa la lista
```

```python
F = funcion(100000) #se manda a llamar a la función con 100000 iteraciones
```

```python
plt.hist(F,bins = 100)   #mostramos el histograma
plt.show()
```
