```python
from prettytable import PrettyTable

elementos = int(input('Ingrese cuántos datos son: '))
Significancia = float(input('Ingresa el valor de significancia de la tabla ji-cuadrada: '))
#categoria = int(input ('Ingrese el número de clases'))
Lista=[]
for i in range(elementos):
  elemento_i = float(input(f'Ingresa el dato {i+1}: '))
  Lista.append(elemento_i)
a=0
b=0
c=0
d=0
e=0
#establecemos los intervalos
for i in Lista:
  if 0 <= i < 0.2:
    a = a+1
  elif 0.2 <= i < 0.4:
    b = b+1
  elif 0.4 <= i < 0.6:
    c = c+1
  elif 0.6 <= i < 0.8:
    d = d+1
  elif 0.8 <= i < 1.0:
    e = e+1

Intervalos={'1': '(0.0-0.1999)',
            '2': '(0.2-0.3999)',
            '3': '(0.4-0.5999)',
            '4': '(0.6-0.7999)',
            '5': '(0.8-0.9999)'}
Oi={'1': a,
    '2': b,
    '3': c,
    '4': d,
    '5': e }

print('\n')
X2=0
Frecuencia_esperada = elementos/5
Tabla = PrettyTable(['Intervalo','fo','fe','fo-fe','(fo-fe)^2 / fe'])

for i in Intervalos:
  diferencia=Oi[i]-Frecuencia_esperada
  cuadrado_div = (diferencia**2)/Frecuencia_esperada
  X2=X2+cuadrado_div

  listabla = [Intervalos[i],Oi[i],Frecuencia_esperada,diferencia,cuadrado_div]
  Tabla.add_row(listabla)

print(Tabla)
print('\n')
print('El valor de ji-cuadrada es: ',X2,'\n')

if X2<=Significancia:
  print('Como ',X2,' <= ',Significancia)
  print('No se rechaza la hipotesis Ho')
else:
  print('Como ',X2,' > ',Significancia)
  print('Se rechaza Ho')
```
