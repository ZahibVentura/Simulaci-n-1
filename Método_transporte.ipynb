```python
def redistribuir(matriz, oferta, demanda):
  filas, columnas = matriz.shape
  while True:
    movido = False
    for i in range(filas):
      for j in range(columnas):
        if matriz[i][j] != 0:
          # Intentar mover hacia la derecha
          if j + 1 < columnas and matriz[i][j + 1] == 0:
            cantidad = min(matriz[i][j], demanda[j])
            matriz[i][j] -= cantidad
            matriz[i][j + 1] += cantidad
            demanda[j] -= cantidad
            movido = True
          # Intentar mover hacia abajo
          elif i + 1 < filas and matriz[i + 1][j] == 0:
            cantidad = min(matriz[i][j], oferta[i])
            matriz[i][j] -= cantidad
            matriz[i + 1][j] += cantidad
            oferta[i] -= cantidad
            movido = True
    if not movido:
      break
  return matriz

# Ejemplo de uso
matriz = np.array([[10, 0, 20], [0, 30, 40]])
oferta = [30, 70]
demanda = [40, 30, 30]
resultado = redistribuir(matriz.copy(), oferta.copy(), demanda.copy())
print("Matriz original:")
print(matriz)
print("Matriz redistribuida:")
print(resultado)
```
