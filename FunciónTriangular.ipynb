```python
import random as rd
import matplotlib.pyplot as plt

def triangular(a=1,b=2,c=4,n=50):
  Lista=[]
  for i in range(n):
    U = rd.uniform(0,1)

    if 0<=U<((b-a)/(c-a)):
      x = a + (U*(c-a)*(b-a))**(1/2)
      Lista.append(x)
    elif ((b-a)/(c-a))<=U<=1:
      x = c - ((1-U)*(c-a)*(c-b))**(1/2)
      Lista.append(x)

  return Lista

A=int(input("Ingrese a: "))
B=int(input("Ingrese b: "))
C=int(input("Ingrese c: "))
Numero=int(input("Ingrese clas iteraciones: "))

f = triangular(A,B,C,Numero)

#plt.plot(f)
#plt.show()
plt.hist(f,bins=100)
plt.show()
```

```python
import random as rd
import matplotlib.pyplot as plt

def distribuciones(n=30):
  Lista=[]
  for i in range(n):
    U = rd.uniform(0,1)

    if 0<=U<(3/4):
      x = (5/4)-((25/16)-2*U)**(1/2)
      Lista.append(x)
    elif (3/4)<=U<=1:
      x = 4*U-2
      Lista.append(x)

  return Lista


Numero=int(input("Ingrese las iteraciones: "))

Y = distribuciones(Numero)

#plt.plot(f)
#plt.show()
plt.hist(Y,bins=100)
plt.show()
```
