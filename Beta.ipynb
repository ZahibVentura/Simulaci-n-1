```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def funny(n=50):
  lista = []
  for i in range(n):
    x = rd.uniform(0,1)
    U=(20*x)*(1-x)**3
    lista.append(U)

  return lista

Iteraciones = int(input("Ingrese las iteraciones: "))

F = funny(Iteraciones)

plt.hist(F, bins=100)
plt.show()
```
