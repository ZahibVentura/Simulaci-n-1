```python
 #prueba Kolmogorov
from prettytable import PrettyTable
n=int(input('Numero de datos: '))
valor_critico = float(input('Ingresa el valor crítico: '))
numeros = [0.8797, 0.3884, 0.6289, 0.8750, 0.5999, 0.8589, 0.9996, 0.2415, 0.3808, 0.9606,
           0.9848, 0.3469, 0.7977, 0.5844, 0.8147, 0.6431, 0.7387, 0.5613, 0.0318, 0.7401,
           0.4557, 0.1592, 0.8536, 0.8846, 0.3410, 0.1492, 0.8681, 0.5291, 0.3188, 0.5992,
           0.9170, 0.2204, 0.5991, 0.5461, 0.5739, 0.3254, 0.0856, 0.2258, 0.4603, 0.5027,
           0.8376, 0.6235, 0.3681, 0.2088, 0.1525, 0.2006, 0.4720, 0.4272, 0.6360, 0.0954]
Lista_ordenada = sorted(numeros)
Lista_Dp = []
Lista_Dm = []
Tabla=PrettyTable(['i','xi','D-','D+'])
for i in range(50):
  Dp=abs((i+1)/n - Lista_ordenada[i])
  Lista_Dp.append(Dp)
  Dm=abs(Lista_ordenada[i] - (i-1+1)/n)
  Lista_Dm.append(Dm)
  Tabla.add_row([i+1,Lista_ordenada[i],Lista_Dm[i],Lista_Dp[i]])

Maxi_Dp = max(Lista_Dp)
Mini_Dm = max(Lista_Dm)
if Maxi_Dp >= Mini_Dm:
  D=Maxi_Dp
else:
  D=Mini_Dm
print('\n')
print('D+: ',Maxi_Dp)
print('D-: ',Mini_Dm)
print(Tabla)
print('El punto crítico es: ',D)
if D<valor_critico:
  print('se acepta la hipotesis')
else:
  print('se rechaza')
```
