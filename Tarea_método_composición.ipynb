```python
import random as rd
import matplotlib.pyplot as plt   #importamos todas las librerias a utilizar
import numpy as np
```

```python
#La siguiente parte del codigo grafica la función la cual se intenta simular

#Puesto que es una funcion a trozos se define el intervalo 8<=X<9 de para la primera parte de la función
a = 8
b = 9
X = np.linspace(a, b, 50)
Y = (2*X-15)/4   #primer trozo de la función

c = 9   #intervalo 9<X<=10
d = 10
V = np.linspace(c,d, 50)
W = (21-2*V)/4  #segundo trozo de la función

x = [8,9,10]
y = [0.25,0.75,0.25]

plt.figure()

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})') #se muestran los puntos donde hay la intersección de las funciones y extremos

plt.plot(X, Y, 'k')   #se grafican ambas funciones por partes (son dos partes)
plt.plot(V, W, 'k')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')  #titulos de ejes
plt.grid()
plt.show()
```

```python
#La siguiente parte del codigo grafica la función acumulativa

X = np.linspace(a, b, 50)
Y = (X**2-15*X+56)/4     #esta es la función acumulativa del primer trozo de la función

V = np.linspace(c,d, 50)
W = 0.5 + (-V**2 + 21*V - 108)/4   #esta es la función acumulativa del segundo trozo de la función

x = [8,9,10]
y = [0,0.5,1]

plt.figure()

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})')  #se muestran los puntos donde hay la intersección de las funciones acumulativas y extremos

plt.plot(X, Y, 'k')
plt.plot(V, W, 'k')  #se grafican ambas funciones acumulativas por partes (son dos partes)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')  #titulos a los ejes
plt.grid()
plt.show()
```

```python
#La función de la grafica es una f.d.p
#Se comprobó que la integral total da como resultado 1
#además la función siempre es positiva para todo X en el intervalo [8,9]
#se aplica el método de la composición para su simulación
# A1 = A2 = (1/2)
#Si u1<=0.5 hacer:
#      x = (15+(8*u2+1)^(0.5))/2
#otro caso hacer:
#      x = (21-(9-8*u2)^(0.5))/2

#El siguiente código muestra la simulación de dicha función

def funcion(N=50): #Definimos una función que regresa una lista de valores x
  Lista = []
  for i in range(N):
    u1 = rd.uniform(0,1) #variable uniforme de 0 a 1
    u2 = rd.uniform(0,1) #variable uniforme de 0 a 1
    if u1 <= 0.5:
      z = (15+(8*u2+1)**(0.5))/2
    else:
      z = (21-(9-8*u2)**(0.5))/2
    Lista.append(z)
  return Lista # se regresa la lista
```

```python
F = funcion(100000) #se manda a llamar a la función con 100000 iteraciones
```

```python
plt.hist(F,bins = 200)   #mostramos el histograma
plt.show()
```
