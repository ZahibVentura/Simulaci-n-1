```python
import random as rnd
import matplotlib.pyplot as plt

No_volados = int(input('Cuantos volados quiere: '))
Lista = []
Resultado = []
Probabilidad_soles = []
Probabilidad_aguilas = []
a=0
s=0
for i in range(No_volados):
  x = rnd.random()
  Lista.append(x)
  if 0 <= x < 0.5:
    a += 1
    p_sol = a/(i+1)
    Resultado.append('Sol')
    Probabilidad_soles.append(p_sol)
  elif 0.5 <= x <=1:
    s += 1
    p_aguila = s/(i+1)
    Resultado.append('Aguila')
    Probabilidad_aguilas.append(p_aguila)

print(Lista)
print(Resultado)
print(Probabilidad_aguilas)
# Crear la gr치fica
plt.plot(Probabilidad_aguilas)

# Mostrar la gr치fica
plt.show()

```
